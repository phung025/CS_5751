#### Part A - Output the boxplots of the attributes bilirubin, alk phosphate, and sgot in one figure

```python
def boxplot(data, col_names):

    dfs = []
    
    # Process the columns to include in boxplot
    for name in col_names:
        df1 = data[[name]]
        df1.columns = ['Y']
        df2 = DataFrame(np.full((len(df1), 1), name), columns=['Name'])
        plot_df = df1.join(df2)
        
        # Drop Nan values for plotting
        plot_df = plot_df.dropna()

        # Convert the column to numeric type
        plot_df['Y'] = plot_df['Y'].apply(pd.to_numeric)
        
        dfs.append(plot_df)
    
    plot_df = pd.concat(dfs)
    ax = sns.boxplot(x="Name", y="Y", data=plot_df)
```

#### Part B - From the boxplots of the three attributes of Task (a), identify which attributes have outliers, and justify your answers. If there are outliers, write code to remove the entire tuples containing the outliers from the dataset and print the dataset after those tuples have been removed.

```python
def non_outliers_bounds(data):
    data = data[~np.isnan(data)]
    data = np.sort(data, kind='quicksort')
    
    q1 = np.median(data[:int(len(data)/2)])
    q3 = np.median(data[int(len(data)/2):])    
    iqr = q3-q1

    return (q1-1.5*iqr, q3+1.5*iqr)
```

```python
tmp = input_data[['Class', 'BILIRUBIN', 'ALK PHOSPHATE', 'SGOT']]
tmp = tmp.apply(pd.to_numeric)

print("Non-outliers boundary of BILIRUBIN", non_outliers_bounds(tmp['BILIRUBIN']))
print("Non-outliers boundary of ALK PHOSPHATE", non_outliers_bounds(tmp['ALK PHOSPHATE']))
print("Non-outliers boundary of SGOT", non_outliers_bounds(tmp['SGOT']))
```

In the figures above, all 3 attributes have outliers. The outliers are the data points that are above or below the whiskers of each boxplot. For example, in the boxplot of Bilirubin, the outliers have values outside of the bound (-0.5, 2.7), the outliers in boxplot of Alk Phosphate are values outside of the bound (-14.5, 221.5), and the outliers in Sgot have values outside the boundary (-73.25, 204.75)

```python
processed_data = input_data[['Class', 'BILIRUBIN', 'ALK PHOSPHATE', 'SGOT']]
processed_data = processed_data.apply(pd.to_numeric)
```

```python
# Remove the rows with outliers WITHOUT touching the row that does not have outlier but 
# contain NaN value
def remove_outliers(data):
    
    headers = processed_data.columns.tolist()[1:]
    bounds = {}
    for header in headers:
        bounds[header] = non_outliers_bounds(data[header])

    # For each row in data
    dfs = []
    for i in range(0, len(data)):
        row = data.iloc[i]
        not_add = False
        
        for header in headers:
            if row[header] is not np.nan and (bounds[header][1] < row[header] or bounds[header][0] > row[header]):
                not_add = True
                break
        
        if not_add == False:
            dfs.append(DataFrame(data.iloc[[i]]))
            
    return pd.concat(dfs)
```

```python
# Remove the outliers from the dataset
d = remove_outliers(processed_data)
```

```python
# Print the data set after removing the outliers
print(d)
```

#### Part C - Replace the missing values of each categorical attribute in the dataset with its mode, and replace the missing values of each continuous attribute in the dataset with its mean, and print the means and modes (according to the nature of the attribute) of each attribute after the missing values have been replaced.

```python
def replace_nan_with_mode(data, continuous_attributes):
    headers = data.columns.tolist()[1:]
    
    for header in headers:
        
        filled_value = -1
        if header in continuous_attributes:
            filled_value = np.mean(data[header])

            # Print the value
            print("Filling column %s with mean %.4f" % (header, filled_value))
        else:
            filled_value = scipy.stats.mode(data[header])[0][0]
            
            # Print the value
            print("Filling column %s with mode %.4f" % (header, filled_value))
        
        # Fill the value to the respected column
        data[header] = data[header].fillna(filled_value)
    
```

#### Part D - Using the preprocessed dataset obtained from Tasks (b) and (c) and using the Cart algorithm, build a decision tree that classifies the tuples based on the class attribute ‘class’ in the dataset. Output the resulting decision tree in textual format. Then evaluate the error rate of the tree using k-fold cross validation sampling with k=10 folds. For each iteration of k-fold CV, print the confusion matrix to standard output, then calculate, print and store the error rate. For this section, you may use any package that implements Cart; you do not need to implement it by yourself.

```python
from sklearn.datasets import load_iris
from sklearn import tree

# Get the input & response matrix
hepatitis_data = d[d.columns.tolist()[1:]].values
hepatitis_classes = d[d.columns.tolist()[0]].values

clf = tree.DecisionTreeClassifier(criterion="entropy", splitter="best", max_depth=3)
clf = clf.fit(hepatitis_data, hepatitis_classes)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

n_nodes = clf.tree_.node_count
children_left = clf.tree_.children_left
children_right = clf.tree_.children_right
feature = clf.tree_.feature
threshold = clf.tree_.threshold
impurity = clf.tree_.impurity
classes = clf.classes_
value = clf.tree_.value

# The tree structure can be traversed to compute various properties such
# as the depth of each node and whether or not it is a leaf.
node_depth = np.zeros(shape=n_nodes, dtype=np.int64)
is_leaves = np.zeros(shape=n_nodes, dtype=bool)
stack = [(0, -1)]  # seed is the root node id and its parent depth
while len(stack) > 0:
    node_id, parent_depth = stack.pop()
    node_depth[node_id] = parent_depth + 1

    # If we have a test node
    if (children_left[node_id] != children_right[node_id]):
        stack.append((children_left[node_id], parent_depth + 1))
        stack.append((children_right[node_id], parent_depth + 1))
    else:
        is_leaves[node_id] = True

print("The decision tree structure has %s nodes and has "
      "the following tree structure:\n"
      % n_nodes)

for i in range(n_nodes):
    if is_leaves[i]:
        print("%snode=%s leaf node: Node impurity %.2f. Class %s" % (node_depth[i] * "\t", 
                                                                     i, 
                                                                     impurity[i], 
                                                                     classes[np.argmax(value[i])]))
    else:
        print("%snode=%s internal node: go to node %s if X[:, %s] <= %.2f else to "
              "node %s. Node impurity %.2f. Class %s"
              % (node_depth[i] * "\t",
                 i,
                 children_left[i],
                 feature[i],
                 threshold[i],
                 children_right[i],
                 impurity[i],
                 classes[np.argmax(value[i])]))
```

```python
from sklearn.metrics import confusion_matrix

def k_fold_cross_validation(input_data, k):
    
    result = []
    rates = []
    
    # Partition the data into k parts
    permuted_indices = np.random.permutation(len(input_data))
    dfs = []
    for i in range(k):
        dfs.append(input_data.iloc[permuted_indices[i::k]])
    
    # Perform K-Fold cross validation
    for i in range(k):
        
        # Get training set
        training_set = []
        for j in range(k):
            if j != i:
                training_set.append(dfs[j])
        training_set = pd.concat(training_set)
        
        training_X = training_set[training_set.columns.tolist()[1:]].values
        training_Y = training_set[training_set.columns.tolist()[0]].values

        # Build the decision tree based on the k-1 partitions
        classifier = tree.DecisionTreeClassifier(criterion="entropy", splitter="best", max_depth=3)
        classifier = classifier.fit(training_X, training_Y)
        
        # Get test set
        test_set = dfs[i]
        test_X = test_set[test_set.columns.tolist()[1:]].values
        test_Y = test_set[test_set.columns.tolist()[0]].values
        
        predicted = classifier.predict(test_X)
        
        print("Predicted class: ", predicted)
        print("Expected class:  ", test_Y)
        print("Confusion matrix:")
        
        conf_matrix = confusion_matrix(test_Y, predicted)
        accuracy = np.trace(conf_matrix) / np.sum(conf_matrix)
        rates.append(1-accuracy)
        
        print(conf_matrix)
        print("Error rate: %.4f" % (1-accuracy))
        print("")
    return rates
        
```

```python
error_rates = k_fold_cross_validation(d, 10)
```

```python
print("The true error rate is approximately %.4f. That is %.4f%% of the test data points are misclassified" % (np.mean(error_rates), np.mean(error_rates)*100))
```
