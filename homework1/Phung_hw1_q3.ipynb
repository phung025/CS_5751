### Part a)

```r
library(GGally)
library(corrplot)
```

```r
# Read the data into a dataframe
df <- read.table("dow_jones_index.data", header=TRUE, sep=",")
```

```r
head(df, n=10)
```

```r
# Take out the columns with categorical values
nums <- sapply(df, is.numeric)
df <- df[,nums]

# Take out rows contain NA value
df <- na.omit(df)
```

```r
GGally::ggpairs(df, cardinality_threshold=NULL)
```

The pair of 2 variables "volume" and "previous_weeks_volume" exhibits a strong linear relationship, which indicates a high correlation.

### Part b)

```r
corrplot.mixed(cor(df), upper="color", tl.pos="lt", lower.col="black")
```

The plot shows the correlation between pair of variables in the data frame. This plot indicates that there is a high correlation between 2 variables "volume" and "previous_weeks_volume". The correlation between these 2 variables has a value of 0.89
