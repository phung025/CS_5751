```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
def convex(x, y):
    return (x**2)+(y**2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-4, 4, 0.02)
X, Y = np.meshgrid(x, y)
zs = np.array([convex(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

convex_surface = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(0, 50)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Set axes labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a color bar which maps values to colors.
fig.colorbar(convex_surface, shrink=0.5, aspect=5)

# Plot the vector field
x, y, z = np.meshgrid(np.arange(-4, 4, 1),
                      np.arange(-4, 4, 1),
                      np.arange(0, 0.5, 0.5))

u = 2*x
v = 2*y
w = 0

# Set figure title
plt.title(r'Plot of $z=x^2+y^2$ and its gradient')

# Plot vector field
ax.quiver(x, y, z, u, v, w, length=0.1)

plt.show()
```
