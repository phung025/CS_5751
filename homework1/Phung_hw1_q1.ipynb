```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import seaborn as sns
import math
import matplotlib.pyplot as plt

%matplotlib inline
```

### Part a)

$$f(x_0+h)=f(x_0)+\frac{f'(x_0)(h-x_0)}{1!}+\frac{f'(x_0)(h-x_0)^2}{2!}+\frac{f'(x_0)(h-x_0)^3}{3!}+\frac{f'(x_0)(h-x_0)^4}{4!}+\frac{f'(x_0)(h-x_0)^5}{5!}$$

### Part b)

```python
def taylor1(h):
    x = 0
    return np.sin(x) + np.cos(x)*h

def taylor3(h):
    x = 0
    return np.sin(x) + np.cos(x)*h + (-np.sin(x)*(h**2)/2) + (-np.cos(x)*(h**3)/6)

def taylor5(h):
    x = 0
    return np.sin(x) + np.cos(x)*h + (-np.sin(x)*(h**2)/2) + (-np.cos(x)*(h**3)/6) + (np.sin(x)*(h**4)/24) + (np.cos(x)*(h**5)/120)

def sin(x):
    return np.sin(x)

t1 = np.arange(-4, 4, 0.1)
t2 = np.arange(-4, 4, 0.1)
t3 = np.arange(-4, 4, 0.1)
t4 = np.arange(-4, 4, 0.1)

# Label the title
plt.title(r"Taylor Series Expansion of $sin(x)$ Around $x_0=0$")

# Label the axes
plt.ylabel("y")
plt.xlabel("x")

# Plot the series and the legend
plt.plot(t1, taylor1(t1), 'orange', label='Taylor Series (Order 1)')
plt.plot(t2, taylor3(t2), 'g', label='Taylor Series (Order 3)')
plt.plot(t3, taylor5(t3), 'r', label='Taylor Series (Order 5)')
plt.plot(t4, sin(t4), 'b', label='sin(x)')
plt.legend(loc='upper right')
```

### Part c)

Using a higher order Taylor Series, the plot of the series starts to get closer to the plot of sin(x). This is because with a higher order series, we have a more accurate approximation of the function sin(x)
